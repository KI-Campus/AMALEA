# Was ist ein Notebook?
Das Dokument, das Sie lesen, ist keine statische Webseite, sondern eine interaktive Umgebungin der Sie Code schreiben und ausführen können.
Jupyter Notebooks - kurz Notebooks - bestehen aus Zellen. Diese Zellen können eine von zwei Formen annehmen: Text oder Code.

## Codezeile
Hier ist beispielsweise eine <strong>Codezelle</strong> mit einem kurzen Python-Skript, das einen Wert berechnet, ihn in einer Variablen speichert und das Ergebnis ausgibt.


```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Wählen Sie zum Ausführen des Codes in der Zelle oben die Zelle mit einem Klick aus. 

In Google Colab drücken Sie dann entweder die `▶️`-Schaltfläche zum Abspielen links neben dem Code.
Bei Lokaler Installation oder mybinder muss die Zelle zuerst ausgewählt werden, z.B. durch einen Mausklick auf die Zelle. Anschließend kann mittels `▶️ Run`oder `Run Cells`, welches unter `Cell` in der Menüleiste zu finden ist, die Zelle ausgeführt werden. 

Alternativ verwenden Sie die Tastenkombination "Befehlstaste/Strg + Enter". 
Klicken Sie zum Bearbeiten des Codes einfach auf die Zelle und beginnen Sie mit der Bearbeitung.


Codezellen bestehen aus drei Bereichen: dem Eingabebereich, dem Anzeigebereich und dem Ausgabebereich. Der Eingabebereich wird durch die Eingabeaufforderung In []: auf der linken Seite der Zelle gekennzeichnet. Zwischen den Klammern der Eingabeaufforderung In kann eines von drei Elementen stehen: eine Zahl, ein Sternchen oder ein Leerzeichen. Eine Zahl zeigt an, dass diese Zelle ausgeführt wurde, und der Wert der Zahl gibt die Reihenfolge der Ausführung an. Wenn Sie zum Beispiel die erste Zelle nach dem Öffnen eines Notizbuchs ausführen, lautet die Eingabeaufforderung normalerweise In [1]:.

Bevor eine Zelle ausgeführt wird, ist der Bereich der Eingabeaufforderung leer. Daher können Sie auf einen Blick erkennen, dass diese Zelle noch nicht ausgeführt wurde. Wenn eine Eingabeaufforderung eine Nummer enthält, kann es auch sein, dass die Zelle in der Vergangenheit ausgeführt worden ist. Es kann jedoch sein, dass die Zelle in dieser Sitzung noch nicht ausgeführt wurde und die Ausgabe daher möglicherweise alte Ergebnisse anzeigt. 

Während der Ausführung einer Zelle wird die Eingabeaufforderung ein Sternchen enthalten. Wenn anscheinend zu viel Zeit verstrichen ist und Sie immer noch In [*] sehen: Ihr Code befindet sich möglicherweise in einer Endlosschleife, oder Sie haben die Kommunikation mit dem Kernel verloren. Möglicherweise müssen Sie den Kernel unterbrechen oder neu starten. 

Schließlich ist es wichtig, den Anzeige- und den Ausgabebereich unterhalb der Eingabezelle getrennt zu halten. Der Unterschied zwischen diesen beiden Bereichen ist subtil und verwirrend, aber in manchen Fällen sehr wichtig. Der Anzeigebereich ist für alle Elemente reserviert, die der Code zur Anzeige produziert hat. Dazu gehört einfacher Text (z. B. `print("Hallo, Welt")`) oder Zahlen aus einer Darstellung. Der Ausgabebereich ist für Elemente reserviert, die die Zelle "zurückgibt". Aus diesem Grund sehen Sie in vielen Notizbüchern eine Variablenzuweisung, auf die sofort die Variable folgt, wie im Beispiel oben.


Variablen, die Sie in einer Zelle definieren, können später in anderen Zellen verwendet werden:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Mit Notebooks können Sie <strong>ausführbaren Code</strong> und <strong>Rich Text</strong> in einem einzigen Dokument gemeinsam mit <strong>Bildern</strong>, <strong>HTML</strong>, <strong>LaTeX</strong> und vielem mehr kombinieren. 

## Data Science

Sie können das volle Potenzial der beliebten Python-Bibliotheken nutzen, um Daten zu analysieren und visualisieren. Die Codezelle unten verwendet <strong>NumPy</strong>, um zufällige Daten zu erstellen, und <strong>Matplotlib</strong>, um sie zu visualisieren. Klicken Sie zum Bearbeiten des Codes einfach auf die Zelle und beginnen Sie mit der Bearbeitung. Oder Sie führen Sie einfach aus

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
